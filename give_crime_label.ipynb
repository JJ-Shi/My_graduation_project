```python
#Label the 2019 Part B dataset
data_Part_B_df_2019['Exclusion'] = 0
data_Part_B_df_2019['Rndrng_NPI'] = data_Part_B_df_2019['Rndrng_NPI'].astype('int')
for row in tqdm(data_Part_B_df_2019.itertuples()):
    if getattr(row, 'Rndrng_NPI') in crime_data_df_2019_list:
        data_Part_B_df_2019.loc[getattr(row, 'Index'), 'Exclusion'] = 1
```

```python
#Label the 2017 Part B dataset
data_Part_B_df_2017['Exclusion'] = 0
data_Part_B_df_2017['Rndrng_NPI'] = data_Part_B_df_2017['Rndrng_NPI'].astype('int')
for row in tqdm(data_Part_B_df_2017.itertuples()):
    if getattr(row, 'Rndrng_NPI') in crime_data_df_2017_list:
        data_Part_B_df_2017.loc[getattr(row, 'Index'), 'Exclusion'] = 1
```

```python
#Sum the number of successfully labeled 2017 data
Number = 0
for row in tqdm(data_Part_B_df_2017.itertuples()):
    if getattr(row, 'Exclusion') == 1:
        Number = Number +1
        
Number
```
