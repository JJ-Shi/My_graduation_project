```python
import pandas as pd
import math
from tqdm import tqdm

#log the LEIE dataset
crime_col_names =['LASTNAME','FIRSTNAME','MIDNAME','BUSNAME','GENERAL','SPECIALTY','UPIN','NPI','DOB','ADDRESS','CITY','STATE','ZIP','EXCLTYPE','EXCLDATE','REINDATE','WAIVERDATE','WVRSTATE']
crime_data = pd.read_csv('LEIE\LEIE.csv', header = None, names = crime_col_names)
crime_data = pd.DataFrame(crime_data).loc[1:].reset_index(drop=True)

crime_data
```

```python
#Extract data on crimes committed in 2017 and later
crime_data_2017 = []
nan = str('nan')
zero = '0000000000'

for row in tqdm(crime_data.itertuples()):
    if str(getattr(row, 'EXCLDATE')) != nan:
        if int(str(getattr(row, 'EXCLDATE'))[0:4]) >= 2017:
            if getattr(row, 'NPI') != zero:
                crime_data_2017.append(row)
                
crime_data_2017 = pd.DataFrame(crime_data_2017)
```

```python
crime_data_2017
```

```python
#Save the generated crime 2017 data
crime_data_2017.to_csv('crime_data/crime_Data_2017.csv',index = False)
```
